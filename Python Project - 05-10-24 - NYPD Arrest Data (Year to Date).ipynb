```python
#Dropping rows that have null/missing values in the 'KY_CD' column
arrestdf = arrestdf.dropna(subset=['KY_CD'])
```

```python
#Reconfirming the null values were dropped from the 'LAW_CAT_CD' column
arrestdf['LAW_CAT_CD'].unique()
```

```python
# Filtering DataFrame to see rows where the column 'LAW_CAT_CD' has a value of '9' or 'I'
filtered_df = arrestdf[(arrestdf['LAW_CAT_CD'] == '9') | (arrestdf['LAW_CAT_CD'] == 'I')]

print(filtered_df)
```

```python
#Dropping Rows that contain '9' or 'I' under the 'LAW_CAT_CD' column. 
arrestdf.drop(arrestdf[(arrestdf['LAW_CAT_CD'] == '9') | (arrestdf['LAW_CAT_CD'] == 'I')].index, inplace=True)
```

```python
#Confirming that values of '9' and 'I' have been removed
arrestdf['LAW_CAT_CD'].unique()
```

```python
#Renaming the letter/initial values in the 'LAW_CAT_CD' column with their full name.
law_cat_cd_full_name = {'F': 'FELONY', 'M': 'MISDEMEANOR', 'V': 'VIOLENT'}

arrestdf['LAW_CAT_CD'] = arrestdf['LAW_CAT_CD'].replace(law_cat_cd_full_name)
```

```python
#Renaming the letter/initial values in the 'ARREST_BORO' column with their full name.
boro_full_name = {'M': 'MANHATTAN', 'K': 'BROOKLYN', 'Q': 'QUEENS', 'B': 'BRONX', 'S': 'STATEN ISLAND'}

arrestdf['ARREST_BORO'] = arrestdf['ARREST_BORO'].replace(boro_full_name)
```

```python
#Defining a function and using "if" and "if-else" statments to link the 'JURISDICTION_CODE' column values to alternative values to be placed in new column called 'JURISDICTION_CODE_DESC'
def jurisdiction_code_desc(jurisdiction_code):
    if jurisdiction_code == 0:
        return 'NYPD PATROL'
    elif jurisdiction_code == 1:
        return 'NYPD TRANSIT'
    elif jurisdiction_code == 2:
        return 'NYPD HOUSING'
    else:
        return 'NON-NYPD'

#Applying code to create the new column called 'JURISDICTION_CODE_DESC'
arrestdf['JURISDICTION_CODE_DESC'] = arrestdf['JURISDICTION_CODE'].apply(jurisdiction_code_desc)
```

```python
#Changing the postion of the 'JURISDICTION_CODE_DESC' column
arrestdf.insert(11, 'JURISDICTION_CODE_DESC', arrestdf.pop('JURISDICTION_CODE_DESC'))
```

```python
#Dropping the last 5 columns in the DataFrame

#Using "len" to get the number of columns in the DataFrame
num_columns = len(arrestdf.columns)

#Using iloc to remove the last 5 columns
arrestdf = arrestdf.iloc[:, :num_columns - 5]
```

# Data Visualization

```python
#Creating a bar graph of the grouped data of "arrests_by_race_borough" to show all the results on 1 graph
plot.figure(figsize=(10, 10))
arrests_by_race_borough.plot(kind='bar', color='maroon')

plot.xlabel('Borough/Race')
plot.ylabel('Arrest Made')
plot.title('Arrest Made for Each Borough by Race')

plot.xticks(rotation=90)
plot.tight_layout()
plot.show()
```

```python
#Creating a stacked bar graph of the grouped data of "arrests_by_race_borough" to show all the results on 1 graph 
arrests_by_race_borough = arrestdf.groupby(['ARREST_BORO', 'PERP_RACE']).size().reset_index(name='Arrest_Count')

#Defining colors for each borough
colors = {'BRONX': 'orange', 'BROOKLYN': 'black', 'MANHATTAN': 'green', 'QUEENS': 'maroon', 'STATEN ISLAND': 'yellow'}

#Creating a stacked bar graph
plot.figure(figsize=(12, 6))

for borough, color in colors.items():
    data = arrests_by_race_borough[arrests_by_race_borough['ARREST_BORO'] == borough]
    plot.bar(data['PERP_RACE'], data['Arrest_Count'], color=color, label=borough)

plot.xlabel('Race')
plot.ylabel('Number of Arrests')
plot.title('Number of Arrests by Race and Borough')
plot.legend()
plot.xticks(rotation=90)
plot.show()
```
