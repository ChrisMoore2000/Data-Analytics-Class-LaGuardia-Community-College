# <h1 style="text-align: center; color: red;">Visualization of Regents Exam Mean Scores for William E. Grady High School</h1>

```python
Grady_df.groupby(['Regents Exam'])['Mean Score'].mean() 
```

```python
Grady_Data = Grady_df
df = pd.DataFrame(Grady_Data)
```

```python
Grady_mean_scores = Grady_df.groupby(['Regents Exam'])['Mean Score'].mean() 
```

```python
plot.figure(figsize=(10, 6))
Grady_mean_scores.plot(kind='bar', color='maroon')

plot.xlabel('Regents Exam')
plot.ylabel('Mean Score')
plot.title('Grady Mean Test Scores by Regents Exam')

plot.xticks(rotation=90)
plot.tight_layout()
plot.show()
```

# <h1 style="text-align: center; color: red;">Visualization of Regents Exam Mean Scores for Abraham Lincoln High School</h1>

```python
Lincoln_df.groupby(['Regents Exam'])['Mean Score'].mean() 
```

```python
Lincoln_Data = Lincoln_df
df = pd.DataFrame(Lincoln_Data)
```

```python
Lincoln_mean_scores = Lincoln_df.groupby(['Regents Exam'])['Mean Score'].mean() 
```

```python
plot.figure(figsize=(10, 6))
Lincoln_mean_scores.plot(kind='bar', color='blue')

plot.xlabel('Regents Exam')
plot.ylabel('Mean Score')
plot.title('Lincoln Mean Test Scores by Regents Exam')

plot.xticks(rotation=90)
plot.tight_layout()
plot.show()
```
