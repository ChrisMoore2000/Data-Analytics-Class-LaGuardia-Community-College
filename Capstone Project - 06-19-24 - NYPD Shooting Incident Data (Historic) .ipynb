```python
#Renaming various columns for better readability and better aesthetics
shooting_incidents_df.rename(columns={
    'INCIDENT_KEY': 'Incident Key',
    'OCCUR_DATE': 'Incident Date',
    'OCCUR_TIME': 'Incident Time',
    'BORO': 'Borough',
    'LOC_OF_OCCUR_DESC': 'Incident Location (Description)',
    'PRECINCT': 'Precinct',
    'JURISDICTION_CODE': 'Jurisdiction Code',
    'LOC_CLASSFCTN_DESC': 'Incident Location (Classification)',
    'LOCATION_DESC': 'Incident Location',
    'STATISTICAL_MURDER_FLAG': 'Statistical Murder Flag',
    'PERP_AGE_GROUP': 'Perp Age Group',
    'PERP_SEX': 'Perp Sex',
    'PERP_RACE': 'Perp Race',
    'VIC_AGE_GROUP': 'Victim Age Group',
    'VIC_SEX': 'Victim Sex',
    'VIC_RACE': 'Victim Race'
}, inplace=True)

#Displaying the column names to verify the changes
print("Column Names:", shooting_incidents_df.columns.tolist())
```

# Changing the Case of Column Values to Title Case

```python
#Changing the case of column values to title case
columns_to_title_case = ['Borough', 'Incident Location (Description)', 'Incident Location (Classification)', 'Incident Location', 'Perp Race', 'Victim Race']

#Applying the title case changes to the specified columns
for columns in columns_to_title_case:
    shooting_incidents_df[columns] = shooting_incidents_df[columns].str.title()
    
#Verifying the title case changes were successfully implemented
shooting_incidents_df.columns
```

# Creating a New Columns for Month and Year

```python
#Converting the 'Incident Date' column to datetime format
shooting_incidents_df['Incident Date'] = pd.to_datetime(shooting_incidents_df['Incident Date'], format='%m/%d/%Y')

#Extracting the 'Month' and 'Year'
shooting_incidents_df['Month'] = shooting_incidents_df['Incident Date'].dt.month
shooting_incidents_df['Year'] = shooting_incidents_df['Incident Date'].dt.year

#Mapping the numerical month values to their respective month names
month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
shooting_incidents_df['Month'] = shooting_incidents_df['Month'].map(month_mapping)

#Converting the 'Year' column to atring
shooting_incidents_df['Year'] = shooting_incidents_df['Year'].astype(str)

#Displaying the DataFrame to verify the changes
shooting_incidents_df.head()
```

# Creating a New Column for Day of the Week

```python
#Extracting the 'Day of the Week'
shooting_incidents_df['Day of the Week'] = shooting_incidents_df['Incident Date'].dt.dayofweek

#Mapping the numerical day values to their respective day names
day_mapping = {
    0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday',
    4: 'Friday', 5: 'Saturday', 6: 'Sunday'
}
shooting_incidents_df['Day of the Week'] = shooting_incidents_df['Day of the Week'].map(day_mapping)

#Displaying the DataFrame to verify the changes
shooting_incidents_df.head()
```

# Creating a New Column for Time of Day

```python
#Converting the 'Incident Time' column to datetime format
shooting_incidents_df['Incident Time'] = pd.to_datetime(shooting_incidents_df['Incident Time'], format='%H:%M:%S').dt.time

#Creating e a new column with 12-hour format with AM/PM
shooting_incidents_df['Incident Time (12 Hour)'] = pd.to_datetime(shooting_incidents_df['Incident Time'].astype(str)).dt.strftime('%I:%M:%S %p')

#Defining a function to determine 'Time of Day'
def get_time_of_day(time):
    if time >= pd.to_datetime('00:00:00').time() and time < pd.to_datetime('06:00:00').time():
        return 'Midnight/Early Morning'
    elif time >= pd.to_datetime('06:00:00').time() and time < pd.to_datetime('12:00:00').time():
        return 'Morning'
    elif time >= pd.to_datetime('12:00:00').time() and time < pd.to_datetime('18:00:00').time():
        return 'Afternoon'
    else:
        return 'Evening'

#Applying the function to create the 'Time of Day' column
shooting_incidents_df['Time of Day'] = shooting_incidents_df['Incident Time'].apply(get_time_of_day)

#Displaying the DataFrame to verify the changes
shooting_incidents_df.head()
```

# Data Visualization

# Creating a Line Graph for Each Year (All Boroughs) - Total Incidents

```python
#Grouping by Borough and Year
shooting_incidents_boroughs_by_year_df = shooting_incidents_df.groupby(['Borough', 'Year']).size().reset_index(name='Incident Count')

#Pivoting the table to have years as columns and boroughs as rows
shooting_incidents_boroughs_by_year_df_pivot = shooting_incidents_boroughs_by_year_df.pivot(index='Year', columns='Borough', values='Incident Count')

#Plotting the graph
plot.figure(figsize=(12, 6))
for borough in shooting_incidents_boroughs_by_year_df_pivot.columns:
    plot.plot(shooting_incidents_boroughs_by_year_df_pivot.index, shooting_incidents_boroughs_by_year_df_pivot[borough], marker='o', label=borough)

plot.title('Total Incidents by Borough and Year')
plot.xlabel('Year')
plot.ylabel('Incident Count')
plot.grid(True)

#Creating the legend
plot.legend(title='Borough', bbox_to_anchor=(1.05, 1), loc='upper left')
plot.tight_layout()

plot.show()
```

# Creating a Line Graph for Each Year (Brooklyn vs Other Boroughs) - Total Incidents

```python
#Grouping by Borough and Year
shooting_incidents_boroughs_by_year_df = shooting_incidents_df.groupby(['Borough', 'Year']).size().reset_index(name='Incident Count')

#Separating th Brooklyn data
brooklyn_data = shooting_incidents_boroughs_by_year_df[shooting_incidents_boroughs_by_year_df['Borough'] == 'Brooklyn']

#Separating the other boroughs data
other_boroughs_data = shooting_incidents_boroughs_by_year_df[shooting_incidents_boroughs_by_year_df['Borough'] != 'Brooklyn']

#Summing the incidents for other boroughs for each year
other_boroughs_sum = other_boroughs_data.groupby('Year')['Incident Count'].sum().reset_index()

#Plotting the graph
plot.figure(figsize=(12, 6))

#Plotting the Brooklyn data
plot.plot(brooklyn_data['Year'], brooklyn_data['Incident Count'], marker='o', label='Brooklyn')

#Plotting the other boroughs data
plot.plot(other_boroughs_sum['Year'], other_boroughs_sum['Incident Count'], marker='o', label='Other Boroughs')

plot.title('Total Incidents in Brooklyn vs. Other Boroughs Over the Years')
plot.xlabel('Year')
plot.ylabel('Incident Count')
plot.grid(True)

#Creating the legend
plot.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plot.tight_layout()

plot.show()
```
