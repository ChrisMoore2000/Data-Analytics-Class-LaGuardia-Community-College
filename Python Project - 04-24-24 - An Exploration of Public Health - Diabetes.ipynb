<h3 align="center">I decided to compare males and females across 3 different age groups (50-60, 60-70, and 70-80). I will compare the averages of the groups in regards to time in hospital and number of lab procedures.

```python
# Creating new DataFrames for Men for the different age groups. (See the next 3 cells)
```

```python
Men_Ages_50_to_60_df = diabetesdf[(diabetesdf['gender']=='Male') & (diabetesdf['age']=='[50-60)')]
```

```python
Men_Ages_60_to_70_df = diabetesdf[(diabetesdf['gender']=='Male') & (diabetesdf['age']=='[60-70)')]
```

```python
Men_Ages_70_to_80_df = diabetesdf[(diabetesdf['gender']=='Male') & (diabetesdf['age']=='[70-80)')]
```

```python
#Confirming the new Dataframes are pulling the correct information.
```

```python
print(Men_Ages_70_to_80_df)
```

```python
# Creating new dataframes for Women for the different age groups. (See the next 3 cells)
```

```python
Women_Ages_50_to_60_df = diabetesdf[(diabetesdf['gender']=='Female') & (diabetesdf['age']=='[50-60)')]
```

```python
Women_Ages_60_to_70_df = diabetesdf[(diabetesdf['gender']=='Female') & (diabetesdf['age']=='[60-70)')]
```

```python
Women_Ages_70_to_80_df = diabetesdf[(diabetesdf['gender']=='Female') & (diabetesdf['age']=='[70-80)')]
```

```python
#Filtering the data for the Men and Women Groups for Average Time in the Hospital. (Using Mean)

Men_Ages_50_to_60_Average_Time_in_Hospital = Men_Ages_50_to_60_df['time_in_hospital'].mean()
Men_Ages_60_to_70_Average_Time_in_Hospital = Men_Ages_60_to_70_df['time_in_hospital'].mean()
Men_Ages_70_to_80_Average_Time_in_Hospital = Men_Ages_70_to_80_df['time_in_hospital'].mean()
Women_Ages_50_to_60_Average_Time_in_Hospital = Women_Ages_50_to_60_df['time_in_hospital'].mean()
Women_Ages_60_to_70_Average_Time_in_Hospital = Women_Ages_60_to_70_df['time_in_hospital'].mean()
Women_Ages_70_to_80_Average_Time_in_Hospital = Women_Ages_70_to_80_df['time_in_hospital'].mean()
```

```python
#Printing the outputs of the filtered codes listed in cell 46.

print({Men_Ages_50_to_60_Average_Time_in_Hospital})
print({Men_Ages_60_to_70_Average_Time_in_Hospital})
print({Men_Ages_70_to_80_Average_Time_in_Hospital})
print({Women_Ages_50_to_60_Average_Time_in_Hospital})
print({Women_Ages_60_to_70_Average_Time_in_Hospital})
print({Women_Ages_70_to_80_Average_Time_in_Hospital})
```

```python
#Filtering the data for the Men and Women Groups for Average Number of Procedures. (Using Mean)

Men_Ages_50_to_60_Average_Num_of_Procedures = Men_Ages_50_to_60_df['num_procedures'].mean()
Men_Ages_60_to_70_Average_Num_of_Procedures = Men_Ages_60_to_70_df['num_procedures'].mean()
Men_Ages_70_to_80_Average_Num_of_Procedures = Men_Ages_70_to_80_df['num_procedures'].mean()
Women_Ages_50_to_60_Average_Num_of_Procedures = Women_Ages_50_to_60_df['num_procedures'].mean()
Women_Ages_60_to_70_Average_Num_of_Procedures = Women_Ages_60_to_70_df['num_procedures'].mean()
Women_Ages_70_to_80_Average_Num_of_Procedures = Women_Ages_70_to_80_df['num_procedures'].mean()
```

```python
#Printing the outputs of the filtered codes listed in cell 50.

print({Men_Ages_50_to_60_Average_Num_of_Procedures})
print({Men_Ages_60_to_70_Average_Num_of_Procedures})
print({Men_Ages_70_to_80_Average_Num_of_Procedures})
print({Women_Ages_50_to_60_Average_Num_of_Procedures})
print({Women_Ages_60_to_70_Average_Num_of_Procedures})
print({Women_Ages_70_to_80_Average_Num_of_Procedures})
```

# <h1 style="text-align: center; color: red; margin-top: 0;" >Data Visualization</h1>

```python
#Creating a Bar Graph for the Men and Women Groups for Average Time in the Hospital

plot.figure(figsize=(14, 6))

colors = ['blue', 'red', 'blue', 'red','blue', 'red']

plot.bar(['Males: 50 to 60', 'Females: 50 to 60', 'Males: 60 to 70', 'Females: 60 to 70', 'Males: 70 to 80', 'Females: 70 to 80'], [Men_Ages_50_to_60_Average_Time_in_Hospital,Women_Ages_50_to_60_Average_Time_in_Hospital,Men_Ages_60_to_70_Average_Time_in_Hospital,Women_Ages_60_to_70_Average_Time_in_Hospital,Men_Ages_70_to_80_Average_Time_in_Hospital,Women_Ages_70_to_80_Average_Time_in_Hospital], color=colors)
plot.xlabel('Group')
plot.ylabel('Time in Hospital (Days)')
plot.title('Average Number of Time in the Hospital (Days)')
plot.show()
```

```python
#Creating a Bar Graph for the Men and Women Groups for Average Number of Procedures

plot.figure(figsize=(14, 6))

colors = ['blue', 'red', 'blue', 'red','blue', 'red']

plot.bar(['Males: 50 to 60', 'Females: 50 to 60', 'Males: 60 to 70', 'Females: 60 to 70', 'Males: 70 to 80', 'Females: 70 to 80'], [Men_Ages_50_to_60_Average_Num_of_Procedures,Women_Ages_50_to_60_Average_Num_of_Procedures,Men_Ages_60_to_70_Average_Num_of_Procedures,Women_Ages_60_to_70_Average_Num_of_Procedures,Men_Ages_70_to_80_Average_Num_of_Procedures,Women_Ages_70_to_80_Average_Num_of_Procedures], color=colors)
plot.xlabel('Group')
plot.ylabel('Number of Procedures')
plot.title('Average Number of Procedures')
plot.show()
```
